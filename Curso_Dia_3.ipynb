#Estructuras de datos
Las estructuras de datos son formas de organizar info para trabajar con ellas de manera eficiente

- Listas -> ordenadas, modificadas, permiten duplicados
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no ordenados, sin duplicados
- Diccionarios -> pares clave:valor

Nota:

- Pilas (FILO -> First in, Last out)
- Colas (FIFO -> First in, First out)
- Listas enlazadas (o doblemente enlazadas)

```python
#Union
matutino={20322,22183,123123,53234}
vespertino={20322,11111,2222222,33333}
total_matriculas=matutino.union(vespertino)
print(total_matriculas)
total_matriculas=matutino | vespertino #union pero con |
print(total_matriculas)
```

```python
#Interseccion
ca={1234,121,646,789}
al={1234,1,2,3,4}
ca_al = ca.intersection(al)
ca_al
#intersecicon con &
ca_al = ca & al
ca_al
```

```python
#Diferencia
solo_calculo = ca.difference(al)
print(solo_calculo)
solo_calcuo = ca - al
print(solo_calculo)
```

##Diccionarios
Los diccionarios osn colecciones no ordenadas de pares claves:valor

```python
alumno = {
    #clave     valor
    "nombre": "Eliud Sanchez",
    "edad": 21,
    "carrera": "LCC",
    "semestre": 7,
    "promedio": 88,
    "calificaciones": [100,94,94,97,100]
}
alumno
```

```python
#--------------------------------------HASTA AQUI------------
```

```python
print("El alumno ",alumno["nombre"], "esta estudiando ",alumno["carrera"])
```

```python
#Agregar al diccionario
alumno["area preferida"] = "Ciencia de datos"
alumno
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for aspecto, valor in alumno.items() #devuelve una vista de todos los pares clave:valor del diccionario
print(aspecto,": ", valor)
```

```python
#el metodo .clear() tambien usar aqui
#Tenemos el .get que nos ayuda a obtener el valor (value) de cierta clave (key)
print(alumno.get("nombre"))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
alumno_lista = list(alumno_1)
print(alumno_lista)
print(type(alumno_lista))
```

```python
#El metodo .values() sirve para obtener los values del diccionario
alumno_v = alumno.values()
print(type(alumno_v))
alumno_v
```

```python
if "carrera" in alumno:
  print("Existe carrera")
else:
  print("No existe carrera")

if("dinero") in alumno:
  print("Alumno tiene dinero")
else:
  print("Alumno no tiene dinero")
```

```python
#.copy() puedes copiat todo el diccionario en una nueva variable
#Update -> Sirve para cambiar los valores (valores) (.update())
alumno.update({"semestre":6})
```
