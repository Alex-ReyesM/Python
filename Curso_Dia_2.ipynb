```python
##elif
#Ahora hagamos un programa en el que preguntamos la calificacion del 0 al 100 y queremos saber su equivalente en el sistema de evaluuacion estadounidense
#A+ = 100
#A = 90-99
#B = 80-89

cal=80
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else: #notemos como al ya no haber una condicion, el else cubre todo lo demas
  print("F")
```

```python
#El modulo (%)
#El modulo devuelve el residiuo de una division
#10 % 3 da 1, porque queda de residuo 1

num = 9
if num % 3 ==0 or num % 5 ==0:
  print(f"El numero {num} es multiplo de 3 o de 5")

num = 6 #saber si el numero es par o impar, usando not
if not num % 2 == 0:
  print(f"El numero {num} es impar")
```

##Ciclo FOR
Los llamados ciclos FOR, nos ayuda a realizar acciones durante una cantidad determinada de veces

```python
#Imprimir los numeros del 0 al 4
for i in range(5): #Toma 5 como el numero final, pero inicia siempre en 0
  print(i) #aqui se declara la variable, y se elimina en este mismo "codigo"
```

```python
for num in range(2,11):
  print(num)
```

```python
cadena = "Python"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

```python
#tmb existe el range(inicio, fin, paso)
for i in range(2,21,2):
  print(i)
```

##While

```python
suma = 0
numeros = int(input("Ingresa el numero para sumar"))

while numeros !=0:
  suma += numeros
  numeros = int(input("Ingresa el numero para sumar o 0 para salir"))

print(f"Tu suma es {suma}")
```

```python
#Usar el while como for
num = 0
while num <= 12:
  print(num)
  num += 1
```
